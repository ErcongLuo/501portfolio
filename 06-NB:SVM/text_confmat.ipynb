```python
### This file is to plot the confusion matrices in heatmap format
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import re # data cleaning
import emoji # data cleaning
import nltk # data cleaning
import string # data cleaning
from sklearn.feature_extraction.text import CountVectorizer # data preparation
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix

nltk.download('words')
words = set(nltk.corpus.words.words())

def plot_confusion_matrix_NB(conf_matrix):
    fig, ax = plt.subplots(figsize=(5, 5))
    ax.matshow(conf_matrix, alpha=0.3)
    for i in range(conf_matrix.shape[0]):
        tempi = 'ELECTRIC'
        for j in range(conf_matrix.shape[1]):
            tempj = 'ICE'
            ax.text(x = j, y = i,s=conf_matrix[i, j], va='center', ha='center', size='xx-large')
    plt.xlabel('Predicted Label', fontsize=18)
    plt.ylabel('True Label', fontsize=18)
    plt.title('Confusion Matrix', fontsize=18)
    plt.show()
```

```python
# Categorical NB
print('Categorical NB Confusion Matrix')
plot_confusion_matrix(np.matrix([[873, 0],[0, 102998]]))
print('0 for Electric, 1 for Internal Combustion')
```

```python
# Gaussian NB
print('Gaussian NB Confusion Matrix')
plot_confusion_matrix(np.matrix([[124,749],[102667, 331]]))
print('0 for Electric, 1 for Internal Combustion')
```

```python
# Non-parametric NB
print('Mixed NB Confusion Matrix')
plot_confusion_matrix(np.matrix([[873, 0],[2086, 100912]]))
print('0 for Electric, 1 for Internal Combustion')
```
